## DATAFRAME SONGS

```python
df_raw = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/proyecto_spotify/dataset/songs.csv")
)
df_raw.display()

# convertir la data songs a tabla 
df_raw.write.mode("overwrite") \
    .saveAsTable("proyecto_spotify.bronze_songs")

```
